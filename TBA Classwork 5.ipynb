### Import necessary libraries

```python
import pandas as pd 
from sklearn import preprocessing
import nltk
```

### To read the Arts_Crafts_and_Sewing_5.json file

```python

df = pd.read_json( 'Arts_Crafts_and_Sewing_5.json',lines=True)
df.sample(10)
```

#### The code downloads NLTK's opinion lexicon, prints the total number of words, and lists the first ten positive and negative words from the lexicon for sentiment analysis.

```python

nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon
from nltk.tokenize import word_tokenize

print('Total number of words in opinion lexicon', len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon',
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon',
      opinion_lexicon.negative()[:10])
```

### To create a dictionary which we can use for scoring our review text

```python
# Let's create a dictionary which we can use for scoring our review text
nltk.download('punkt')
df.rename(columns={"reviewText": "text"}, inplace=True)
pos_score = 1
neg_score = -1
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

#### The bing_liu_score function calculates the sentiment score of a given text by tokenizing it, then summing sentiment values from a predefined word dictionary for each word found.

```python
def bing_liu_score(text):
    sentiment_score = 0
    bag_of_words = word_tokenize(text.lower())
    for word in bag_of_words:
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score  
```

#### The following code fills missing values in the 'text' column of the DataFrame df with 'no review', and then applies the bing_liu_score function to each text entry to compute and assign sentiment scores to a new column 'Bing_Liu_Score'.

```python
df['text'].fillna('no review', inplace=True)
df['Bing_Liu_Score'] = df['text'].apply(bing_liu_score)
```

#### The following code displays the first 10 rows of the 'overall', 'text', and 'Bing_Liu_Score' columns from the DataFrame df.

```python
df[['overall',"text", 'Bing_Liu_Score']].head(10)
```

#### The following code calculates and displays the average 'Bing_Liu_Score' for each unique value in the 'overall' column of the DataFrame df, grouping the data by 'overall'.

```python
df.groupby('overall').agg({'Bing_Liu_Score':'mean'})
```
